```python
import pandas as pd
df = pd.read_csv('./crypto_data.csv')
df.head()
```

```python
def fixCol(col):
    spl = col.split('_')
    if len(spl) == 1: return col
    return spl[1] + '_' + spl[0]

currCols = df.columns
newCols = [fixCol(col) for col in currCols]
df = df.reindex(columns=newCols)
df = df.set_index('time')
df.head()
```

```python
# Save the new df
df.to_csv('./new_crypto_data.csv')
```
