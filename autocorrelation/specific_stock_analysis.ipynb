```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta
```

```python
df = pd.read_csv('./crypto_data.csv')
df['time'] = pd.to_datetime(df['time'])
df = df.set_index('time')
df.head()
print("Daterange:", df.index[0], df.index[-1])
```

```python
bins = [-1,10,100,1000,10000,100000,10000000]
binned = pd.cut(df['OSMOUSDT_turnover'], bins=bins)
binned.value_counts(normalize=True)
```

```python
df['OSMOUSDT_turnover'].plot()
plt.ylim(bottom=0, top=40000)
```

```python
# yesterday
start_time = datetime(2024, 11, 30, 10, 0)
today = df.loc[start_time:]
today.head()

today['OSMOUSDT_turnover'].plot()
plt.ylim(bottom=0, top=40000)

today['OSMOUSDT_turnover'].describe()
```
